## Joining Population Size

```python
whr_ay_data_df
```

```python
# population data from world bank (https://data.worldbank.org/indicator/SP.POP.TOTL?end=2022&start=2002&view=chart)
population_fname = "./data/raw/API_SP.POP.TOTL_DS2_en_csv_v2_6011311/API_SP.POP.TOTL_DS2_en_csv_v2_6011311.csv"
```

```python
# Load population data
pop_df = pd.read_csv(population_fname, skiprows=4)

# Drop empty unnamed column
pop_df.drop(columns=pop_df.columns[67], inplace=True)

pop_df
```

```python
iso3_to_iso2 = {c.alpha_3: c.alpha_2 for c in pycountry.countries}
pop_df["iso_alpha_2"] = pop_df["Country Code"].map(iso3_to_iso2)
pop_df
```

```python
pop_country_year_df = pop_df.drop(columns=['Indicator Name', 'Indicator Code'])
melted_df = pd.melt(pop_country_year_df, id_vars=['Country Name', "iso_alpha_2", 'Country Code'], var_name='Year', value_name='Population')
melted_df["Year"] = melted_df["Year"].astype(int)
melted_df
```

```python
whr_ay_data_df = whr_ay_data_df.rename(columns={"year": "Year", "Country name": "Country Name"})
joined_df = melted_df.set_index(['Country Name', 'Year']).join(whr_ay_data_df.set_index(['Country Name', 'Year']), how='inner').reset_index()
```

```python
joined_df[["Country Name", "Country Code", "iso_alpha_2", "Year", "Population", "Life Ladder"]]
```

## Regional Mapping, using inequalities set

```python
regional_mapping_csv = "./data/raw/WID_countries.csv"
```

```python
regional_mapping_df = pd.read_csv(regional_mapping_csv, delimiter=";", index_col="titlename")
regional_mapping_df
```

```python
ll_scores_by_region_df = regional_mapping_df.merge(joined_df, right_on="iso_alpha_2", left_on="alpha2")[
    ["Year", "region", "region2", "alpha2", "Country Code", "shortname", "Population", "Log GDP per capita", "Life Ladder"]]
ll_scores_by_region_df
```

```python
ll_scores_by_region_df.sort_values(by=["region2", "shortname"]).to_csv("./data/processed/WHR_LL_Region_20052022.csv")
```
