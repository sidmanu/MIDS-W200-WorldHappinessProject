```python
import requests
import json
import geopandas
from geodatasets import get_path
```

```python
# Redlined map
path_to_data = '../data/raw/geojson/CASanFrancisco1937.geojson'
gdf = geopandas.read_file(path_to_data)
gdf["area"] = gdf.area
gdf.plot("area", legend=True)
```

```python
# Interactive map of SF
path_to_data = '../data/raw/geojson/SF Find Neighborhoods.geojson'
gdf = geopandas.read_file(path_to_data)
gdf["area"] = gdf.area
gdf.explore("area", legend=False)
```
